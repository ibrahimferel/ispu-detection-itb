```python
def parse_penanggalan_1(x):
  if pd.isna(x):
    return pd.NaT
  x=str(x)
  if "-" in x:
    return pd.to_datetime(x, format="%Y-%m-%d")
  elif "/" in x:
    return pd.to_datetime(x, format="%m/%d/%Y")
  else:
    return pd.NaT

df_2013["tanggal"] = df_2013["tanggal"].apply(parse_penanggalan_1)
df_2013["tanggal"] = df_2013["tanggal"].dt.strftime("%Y-%m-%d")

def parse_penanggalan_2(x):
  if pd.isna(x):
    return pd.NaT
  x=str(x)
  if "-" in x:
    return pd.to_datetime(x, format="%Y-%m-%d")
  elif "/" in x:
    return pd.to_datetime(x, format="%d/%m/%Y")
  else:
    return pd.NaT

df_2016["tanggal"] = df_2016["tanggal"].apply(parse_penanggalan_2)
df_2016["tanggal"] = df_2016["tanggal"].dt.strftime("%Y-%m-%d")

df_2018["tanggal"] = df_2018["tanggal"].apply(parse_penanggalan_2)
df_2018["tanggal"] = df_2018["tanggal"].dt.strftime("%Y-%m-%d")
```

### Feature Engineering

```python
df_ispu_pseudo['tanggal'] = pd.to_datetime(df_ispu_pseudo['tanggal'])
df_libur['tanggal'] = pd.to_datetime(df_libur['tanggal'])

df = df_ispu_pseudo.merge(
    df_libur[['tanggal', 'is_holiday_nasional', 'is_weekend', 'day_name']],
    on='tanggal',
    how='left'
)

# Hari, bulan, tahun
df['day'] = df['tanggal'].dt.day
df['month'] = df['tanggal'].dt.month
df['year'] = df['tanggal'].dt.year
df['day_of_week'] = df['tanggal'].dt.dayofweek  # Senin=0, Minggu=6

# Sin / cos bulanan untuk musiman tahunan
df['sin_month'] = np.sin(2 * np.pi * df['month']/12)
df['cos_month'] = np.cos(2 * np.pi * df['month']/12)

# Sin / cos mingguan untuk pola mingguan
df['sin_week'] = np.sin(2 * np.pi * df['day_of_week']/7)
df['cos_week'] = np.cos(2 * np.pi * df['day_of_week']/7)

df = df.sort_values(["lokasi_spku", "tanggal"])

lag_days = [1, 7, 30]
rolling_days = [7, 30]

for col in ['pm10','so2','co','o3','no2']:

    for lag in lag_days:
        df[f'{col}_lag{lag}'] = (
            df.groupby("lokasi_spku")[col]
              .shift(lag)
        )

    for window in rolling_days:
        g = df.groupby("lokasi_spku")[col]

        df[f'{col}_roll{window}_mean'] = (
            g.rolling(window).mean()
             .reset_index(level=0, drop=True)
        )

        df[f'{col}_roll{window}_max'] = (
            g.rolling(window).max()
             .reset_index(level=0, drop=True)
        )

        df[f'{col}_roll{window}_std'] = (
            g.rolling(window).std()
             .reset_index(level=0, drop=True)
        )

df['pm10_so2_ratio'] = df['pm10'] / (df['so2'] + 1e-6)  # +epsilon supaya tidak dibagi nol
df['co_o3_ratio'] = df['co'] / (df['o3'] + 1e-6)

df = df.dropna()

df.info()
df.head()
```

```python
from sklearn.neural_network import MLPClassifier
from catboost import CatBoostClassifier
from xgboost import XGBClassifier
from sklearn.impute import SimpleImputer
target = 'categori'
drop_cols = [
    target,        # label
    "tanggal",         # index waktu
    "periode_data", # info waktu mentah
    "critical",     # turunan ISPU
    "max"           # dipakai menentukan kategori
]

cat_features = ["lokasi_spku", 'day_name']
num_features = [
    c for c in df.columns
    if c not in drop_cols + cat_features and df[c].dtype in [np.float64, np.int64]
]


X_train = df_train[cat_features + num_features]
y_train = df_train[target]

X_val = df_val[cat_features + num_features]
y_val = df_val[target]

# Create a numeric transformer with imputation
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),
    ('scaler', StandardScaler()) # Optional: add scaling as well
])

preprocess = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), ['lokasi_spku']),
        ('num', numeric_transformer, num_features) # Use the numeric_transformer here
    ]
)

cpu_models = {
    # ===== Linear =====
    "Logistic Regression": LogisticRegression(max_iter=1000),

    # ===== Distance =====
    "KNN": KNeighborsClassifier(n_neighbors=5),

    # ===== Tree =====
    "Decision Tree": DecisionTreeClassifier(random_state=42),
    "Random Forest": RandomForestClassifier(
        n_estimators=300,
        random_state=42),

    # ===== Boosting =====
    "Gradient Boosting": GradientBoostingClassifier(random_state=42)
}

results = []

for name, model in cpu_models.items():
    pipe = Pipeline([
        ('prep', preprocess),
        ('model', model)
    ])

    pipe.fit(X_train, y_train)
    y_pred = pipe.predict(X_val)

    f1 = f1_score(y_val, y_pred, average='weighted')

    results.append({
        'model': name,
        'f1_score': f1
    })

X_train_cb = df_train.drop(columns=drop_cols)
y_train_cb = df_train[target]

X_val_cb = df_val.drop(columns=drop_cols)
y_val_cb = df_val[target]

cat_features_cb = ['lokasi_spku', 'day_name']

cat_model = CatBoostClassifier(
    iterations=800,
    learning_rate=0.05,
    depth=8,
    loss_function="MultiClass",
    eval_metric="MultiClass",
    task_type="CPU",
    auto_class_weights="Balanced",
    random_seed=42,
    verbose=200
)

cat_model.fit(
    X_train_cb,
    y_train_cb,
    cat_features=cat_features_cb,
    eval_set=(X_val_cb, y_val_cb),
    use_best_model=True
)

y_pred_cb = cat_model.predict(X_val_cb)

results.append({
    'model': 'CatBoost (GPU)',
    'f1_score': f1_score(y_val_cb, y_pred_cb, average='weighted')
})

le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_val_enc = le.transform(y_val)

X_train_xgb = preprocess.fit_transform(X_train)
X_val_xgb = preprocess.transform(X_val)

xgb_model = XGBClassifier(
    n_estimators=600,
    learning_rate=0.05,
    max_depth=8,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="multi:softprob",
    eval_metric="mlogloss",
    tree_method="hist",
    predictor="cpu_predictor",
    random_state=42
)

xgb_model.fit(X_train_xgb, y_train_enc)

y_pred_xgb = xgb_model.predict(X_val_xgb)

results.append({
    'model': 'XGBoost (GPU)',
    'f1_score': f1_score(y_val_enc, y_pred_xgb, average='weighted')
})

df_results = (
    pd.DataFrame(results)
    .sort_values('f1_score', ascending=False)
    .reset_index(drop=True)
)

df_results
```

```python
# Suppress warnings agar output bersih
import warnings
warnings.filterwarnings('ignore')

# --- 1. PREPROCESSING (Encoding) ---
# Kita ubah semua kolom object menjadi angka agar aman masuk ke semua model
# Asumsi: X_train, X_val, y_train, y_val sudah ada dari split sebelumnya

# Buat copy agar data asli aman
X_train_enc = X_train.copy()
X_val_enc = X_val.copy()

# Cari kolom kategorikal
cat_cols = X_train_enc.select_dtypes(include=['object']).columns.tolist()
print(f"Meng-encode kolom: {cat_cols}")

# Encode
encoder = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
X_train_enc[cat_cols] = encoder.fit_transform(X_train[cat_cols])
X_val_enc[cat_cols] = encoder.transform(X_val[cat_cols])

# Siapkan TimeSeriesSplit untuk validasi (Mencegah Data Leakage)
tscv = TimeSeriesSplit(n_splits=3)
```

```python
from sklearn.preprocessing import LabelEncoder

# --- 1. BERSIHKAN DATA TARGET (Hapus 'O3') ---
# Cek apakah ada 'O3' di y_train atau y_val dan hapus baris terkait di X dan y
def clean_garbage_labels(X, y):
    # Cari index yang labelnya BUKAN kategori valid (misal 'O3' atau garbage lainnya)
    valid_categories = ['BAIK', 'SEDANG', 'TIDAK SEHAT']
    mask = y.isin(valid_categories)

    print(f"Data awal: {len(y)}")
    X_clean = X[mask].copy()
    y_clean = y[mask].copy()
    print(f"Data setelah cleaning: {len(y_clean)} (Dibuang {len(y) - len(y_clean)} baris sampah)")

    return X_clean, y_clean

# Terapkan pembersihan (Asumsi X_train_enc sudah ada dari langkah sebelumnya)
X_train_clean, y_train_clean = clean_garbage_labels(X_train_enc, y_train)
X_val_clean, y_val_clean = clean_garbage_labels(X_val_enc, y_val)

# --- 2. ENCODE TARGET (String -> Angka) ---
# XGBoost Wajib pakai angka 0, 1, 2...
le = LabelEncoder()
y_train_final = le.fit_transform(y_train_clean)
y_val_final = le.transform(y_val_clean)

print("Kelas yang dikenali:", le.classes_)
# Output harusnya cuma 5 kelas ISPU, tidak ada 'O3' lagi.

# --- 3. UPDATE FUNGSI OPTUNA ---
# Pastikan menggunakan y_train_final (yang sudah angka)

def tune_xgboost(trial):
    params = {
        'objective': 'multi:softmax',
        'num_class': len(le.classes_), # Gunakan panjang kelas dari LabelEncoder
        'tree_method': 'hist',
        'verbosity': 0,
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'subsample': trial.suggest_float('subsample', 0.6, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.6, 1.0),
        'n_estimators': trial.suggest_int('n_estimators', 100, 500), # Kurangi sedikit biar cepat
        'reg_alpha': trial.suggest_float('reg_alpha', 0, 10),
        'reg_lambda': trial.suggest_float('reg_lambda', 0, 10)
    }

    model = xgb.XGBClassifier(**params)

    # PENTING: Gunakan X_train_clean dan y_train_final
    scores = cross_val_score(model, X_train_clean, y_train_final, cv=tscv, scoring='f1_macro')
    return scores.mean()

# Update juga fungsi lainnya untuk menggunakan y_train_final!

# --- JALANKAN ULANG OPTUNA ---
print("Mulai Tuning Ulang...")
study_xgb = optuna.create_study(direction='maximize')
study_xgb.optimize(tune_xgboost, n_trials=20)
```

```python
df_tft = df.copy()
df_tft["tanggal"] = pd.to_datetime(df_tft["tanggal"])
df_tft = df_tft.sort_values(["lokasi_spku","tanggal"]).reset_index(drop=True)

locations = ["DKI1","DKI2","DKI3","DKI4","DKI5"]
targets   = ["pm10","so2","co","o3","no2"]

max_encoder_length = 60
max_prediction_length = 91

import lightning.pytorch as pl
from lightning.pytorch import Trainer

from pytorch_forecasting import TimeSeriesDataSet
from pytorch_forecasting import TemporalFusionTransformer
from pytorch_forecasting.metrics import QuantileLoss
from pytorch_forecasting.data import GroupNormalizer

tft_models = {}
tft_datasets = {}

for loc in locations:

    df_loc = df_tft[df_tft["lokasi_spku"] == loc].copy()
    df_loc = df_loc.sort_values("tanggal")

    # time index khusus lokasi ini
    df_loc["time_idx"] = np.arange(len(df_loc))

    for target in targets:

        print(f"Training TFT -> {loc} | {target}")

        dataset = TimeSeriesDataSet(
            df_loc,
            time_idx="time_idx",
            target=target,

            group_ids=["lokasi_spku"],

            min_encoder_length=max_encoder_length,
            max_encoder_length=max_encoder_length,

            min_prediction_length=max_prediction_length,
            max_prediction_length=max_prediction_length,

            static_categoricals=["lokasi_spku"],

            time_varying_known_reals=[
                "time_idx"
            ],

            time_varying_unknown_reals=[
                target
            ],

            target_normalizer=GroupNormalizer(
                groups=["lokasi_spku"]
            )
        )

        train_loader = dataset.to_dataloader(
            train=True,
            batch_size=32,
            num_workers=0
        )

        tft = TemporalFusionTransformer.from_dataset(
            dataset,
            learning_rate=0.03,
            hidden_size=16,
            attention_head_size=2,
            dropout=0.1,
            hidden_continuous_size=8,
            loss=QuantileLoss(),
            reduce_on_plateau_patience=3
        )

        trainer = Trainer(
            max_epochs=20,
            accelerator="auto",
            enable_checkpointing=False,
            logger=False,
            gradient_clip_val=0.1
        )

        trainer.fit(tft, train_loader)

        tft_models[(loc, target)] = tft
        tft_datasets[(loc, target)] = dataset

```

```python
df_future = df_future.copy()
df_future["tanggal"] = pd.to_datetime(df_future["tanggal"])

all_predictions = []

for loc in locations:

    df_loc_hist = df_tft[df_tft["lokasi_spku"] == loc].copy()
    df_loc_hist = df_loc_hist.sort_values("tanggal")
    df_loc_hist["time_idx"] = np.arange(len(df_loc_hist))

    last_idx = df_loc_hist["time_idx"].max()

    df_loc_future = df_future[df_future["lokasi_spku"] == loc].copy()
    df_loc_future = df_loc_future.sort_values("tanggal")
    df_loc_future["time_idx"] = np.arange(
        last_idx + 1,
        last_idx + 1 + len(df_loc_future)
    )

    for target in targets:

        print(f"Predicting -> {loc} | {target}")

        # build df_full KHUSUS target ini
        df_full = pd.concat(
            [
                df_loc_hist[["tanggal","lokasi_spku","time_idx", target]],
                df_loc_future[["tanggal","lokasi_spku","time_idx"]]
            ],
            ignore_index=True
        )

        # isi NaN target saja
        df_full[target] = df_full[target].ffill().bfill()

        dataset = tft_datasets[(loc, target)]
        model   = tft_models[(loc, target)]

        predict_dataset = TimeSeriesDataSet.from_dataset(
            dataset,
            df_full,
            predict=True,
            stop_randomization=True
        )

        predict_loader = predict_dataset.to_dataloader(
            train=False,
            batch_size=32,
            num_workers=0
        )

        out = model.predict(
            predict_loader,
            return_x=True,
            mode="prediction"
        )

        preds = out[0]
        x     = out[1]

        preds = preds.detach().cpu().numpy()
        decoder_time_idx = x["decoder_time_idx"].detach().cpu().numpy()

        rows = []
        for i in range(preds.shape[0]):
            for j in range(preds.shape[1]):
                rows.append({
                    "lokasi_spku": loc,
                    "time_idx": int(decoder_time_idx[i, j]),
                    target: float(preds[i, j])
                })

        all_predictions.append(pd.DataFrame(rows))

from functools import reduce

df_pred_tft = pd.concat(
    all_predictions,
    ignore_index=True
)

df_pred_tft = (
    df_pred_tft
    .groupby(["lokasi_spku", "time_idx"], as_index=False)
    .first()
)

df_future_all = []

for loc in locations:

    df_loc_hist = df_tft[df_tft["lokasi_spku"] == loc].copy()
    df_loc_hist = df_loc_hist.sort_values("tanggal")
    df_loc_hist["time_idx"] = np.arange(len(df_loc_hist))

    last_idx = df_loc_hist["time_idx"].max()

    df_loc_future = df_future[df_future["lokasi_spku"] == loc].copy()
    df_loc_future = df_loc_future.sort_values("tanggal")
    df_loc_future["time_idx"] = np.arange(
        last_idx + 1,
        last_idx + 1 + len(df_loc_future)
    )

    df_future_all.append(
        df_loc_future[["lokasi_spku","tanggal","time_idx"]]
    )

df_future_index = pd.concat(df_future_all, ignore_index=True)

df_pred_tft = df_pred_tft.merge(
    df_future_index,
    on=["lokasi_spku","time_idx"],
    how="inner"
)

df_pred_tft = df_pred_tft.sort_values(
    ["lokasi_spku","tanggal"]
).reset_index(drop=True)

df_pred_tft
```

```python
# Coba-Coba 12 entry cihuy

def build_forecast_features_like_train(
    df_hist,
    df_forecast,
    df_libur
):
    base_cols = ["pm10", "so2", "co", "o3", "no2"]

    # pastikan datetime
    df_hist = df_hist.copy()
    df_forecast = df_forecast.copy()

    df_hist["tanggal"] = pd.to_datetime(df_hist["tanggal"])
    df_forecast["tanggal"] = pd.to_datetime(df_forecast["tanggal"])

    # ambil hanya kolom penting dari history
    df_hist_base = df_hist[["tanggal", "lokasi_spku"] + base_cols]

    # gabungkan history + forecast
    df_all = pd.concat(
        [df_hist_base, df_forecast],
        ignore_index=True
    )

    df_all = df_all.sort_values(["lokasi_spku", "tanggal"])

    # ===============================
    # kalender
    # ===============================
    df_all = df_all.merge(
        df_libur[["tanggal", "is_holiday_nasional", "is_weekend", "day_name"]],
        on="tanggal",
        how="left"
    )

    df_all["day"] = df_all["tanggal"].dt.day
    df_all["month"] = df_all["tanggal"].dt.month
    df_all["year"] = df_all["tanggal"].dt.year
    df_all["day_of_week"] = df_all["tanggal"].dt.dayofweek

    df_all["sin_month"] = np.sin(2 * np.pi * df_all["month"] / 12)
    df_all["cos_month"] = np.cos(2 * np.pi * df_all["month"] / 12)
    df_all["sin_week"] = np.sin(2 * np.pi * df_all["day_of_week"] / 7)
    df_all["cos_week"] = np.cos(2 * np.pi * df_all["day_of_week"] / 7)

    # ===============================
    # lag & rolling (recursive style)
    # ===============================
    lag_days = [1, 7, 30]
    roll_days = [7, 30]

    df_all = df_all.sort_values(["lokasi_spku", "tanggal"])

    for col in base_cols:
        for lag in lag_days:
            df_all[f"{col}_lag{lag}"] = (
                df_all
                .groupby("lokasi_spku")[col]
                .shift(lag)
            )

        for w in roll_days:
            df_all[f"{col}_roll{w}_mean"] = (
                df_all
                .groupby("lokasi_spku")[col]
                .rolling(w)
                .mean()
                .reset_index(level=0, drop=True)
            )

            df_all[f"{col}_roll{w}_max"] = (
                df_all
                .groupby("lokasi_spku")[col]
                .rolling(w)
                .max()
                .reset_index(level=0, drop=True)
            )

            df_all[f"{col}_roll{w}_std"] = (
                df_all
                .groupby("lokasi_spku")[col]
                .rolling(w)
                .std()
                .reset_index(level=0, drop=True)
            )

    # ===============================
    # ratio
    # ===============================
    df_all["pm10_so2_ratio"] = df_all["pm10"] / (df_all["so2"] + 1e-6)
    df_all["co_o3_ratio"]   = df_all["co"]   / (df_all["o3"]  + 1e-6)

    # ===============================
    # ambil hanya baris forecast
    # ===============================
    key = set(
        zip(df_forecast["tanggal"], df_forecast["lokasi_spku"])
    )

    mask_forecast = df_all.apply(
        lambda x: (x["tanggal"], x["lokasi_spku"]) in key,
        axis=1
    )

    df_feat = df_all.loc[mask_forecast].copy()

    return df_feat

df_forecast_feat = build_forecast_features_like_train(
    df_hist=df,              # dataframe historis kamu (yang dipakai train)
    df_forecast=df_pred_tft, # output Chronos
    df_libur=df_libur
)

# cat_cols = ['lokasi_spku']
# --- 1. Filter Data & Wilayah (Sesuai Kode Lama) ---
valid_wilayah = ['DKI1', 'DKI2', 'DKI3', 'DKI4', 'DKI5']

if 'tanggal' not in df_forecast_feat.columns and 'date' in df_forecast_feat.columns:
    df_forecast_feat['tanggal'] = df_forecast_feat['date'] # Penyesuaian nama kolom jika perlu

df_forecast_feat['tanggal'] = pd.to_datetime(df_forecast_feat['tanggal'])
df_forecast_clean = df_forecast_feat[df_forecast_feat['lokasi_spku'].isin(valid_wilayah)].copy()

# --- FIX: Tambahkan day_name karena hilang saat forecasting ---
df_forecast_clean['day_name'] = df_forecast_clean['tanggal'].dt.day_name()

print(f"Jumlah data yang akan diprediksi: {len(df_forecast_clean)}")


df_forecast_enc = df_forecast_clean.copy()

# Pastikan urutan kolom sesuai dengan encoder
df_forecast_enc[cat_cols] = encoder.transform(df_forecast_clean[cat_cols])

feature_columns = X_train.columns
X_pred = df_forecast_enc[feature_columns]

y_pred_numeric = final_xgb_model.predict(X_pred)

df_forecast_clean['categori'] = le.inverse_transform(y_pred_numeric)

# Buat ID: YYYY-MM-DD_WILAYAH
df_forecast_clean['id'] = (
    df_forecast_clean['tanggal'].dt.strftime('%Y-%m-%d')
    + '_'
    + df_forecast_clean['lokasi_spku']
)

# --- 5. Finalisasi & Save ---
df_submission = df_forecast_clean[['id', 'categori']].sort_values('id')

# Simpan CSV
filename = 'submission_tft_2.csv'
df_submission.to_csv(filename, index=False)

print(f"Berhasil menyimpan file: {filename}")
print("\nContoh 15 baris pertama:")
print(df_submission.head(15))
```
